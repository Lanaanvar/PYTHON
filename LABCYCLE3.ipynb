```python
#question 4
import pickle
from operator import attrgetter
from fpdf import FPDF

class Vehicle:
    def __init__(self, engine_number, model, type, mileage, vendor, registration_number, owner_name):
        self.engine_number = engine_number
        self.model = model
        self.type = type
        self.mileage = mileage
        self.vendor = vendor
        self.registration_number = registration_number
        self.owner_name = owner_name

    def __str__(self):
        return f"Engine Number: {self.engine_number}\n" \
               f"Model: {self.model}\n" \
               f"Type: {self.type}\n" \
               f"Mileage: {self.mileage}\n" \
               f"Vendor: {self.vendor}\n" \
               f"Registration Number: {self.registration_number}\n" \
               f"Owner Name: {self.owner_name}"

class VehicleCollection:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def delete_vehicle(self, engine_number):
        self.vehicles = [v for v in self.vehicles if v.engine_number != engine_number]

    def modify_vehicle(self, engine_number, new_mileage):
        for vehicle in self.vehicles:
            if vehicle.engine_number == engine_number:
                vehicle.mileage = new_mileage
                break

    def display_collection(self):
        for vehicle in self.vehicles:
            print(vehicle)
            print("")

    def sort_by_mileage(self):
        self.vehicles.sort(key=attrgetter('mileage'))

    def export_report(self, filename):
        pdf = FPDF()
        pdf.add_page()

        for vehicle in self.vehicles:
            pdf.set_font("Arial", size=12)
            pdf.cell(0, 10, txt=str(vehicle), ln=True)

        pdf.output(filename)

    def save_collection(self, filename):
        with open(filename, 'wb') as file:
            pickle.dump(self.vehicles, file)

    def load_collection(self, filename):
        with open(filename, 'rb') as file:
            self.vehicles = pickle.load(file)

# Example usage
collection = VehicleCollection()

n=int(input("Enter number of vehicles : "))

for i in range(n):
  print("Enter the details : ")
  d1 = int(input("Engine No  :"))
  d2 = input("Model : ")
  d3 = input("Type  : ")
  d4 = int(input("Mileage : "))
  d5 = input("Vendor : ")
  d6 = int(input("Registration No. :"))
  d7 = input("Owner Name : ")

  vehicle = Vehicle(d1,d2,d3,d4,d5,d6,d7)

  collection.add_vehicle(vehicle)


print("\n******DISPLAY DETAILS*******")
collection.display_collection()

collection.sort_by_mileage()

print("After sorting based on mileage : ")
print("\n")
collection.display_collection()

m1=int(input("Enter engine number to be modified from the above collection : "))
m2=int(input("Enter new mileage : "))

collection.modify_vehicle(m1,m2)
collection.display_collection()

d=int(input("Enter engine number to be deleted from the above collection : "))

collection.delete_vehicle(d)
collection.display_collection()

collection.export_report("vehicle_report.pdf")

collection.save_collection("vehicle_collection.pickle")

loaded_collection = VehicleCollection()
loaded_collection.load_collection("vehicle_collection.pickle")
# loaded_collection.display_collection()
```
