```python
# function to calculate fibonacci series

def fib(n):
  a=0
  b=1
  c=1
  print(f"1\t|\t{b}\t|\t{a}\t|\t{c}")
  for i in range(2,n+1):
    c=a+b
    print(f"{i}\t|\t{a}\t|\t{b}\t|\t{c}")
    a=b
    b=c

x=int(input("Enter the number of months : "))
print("\n")
print("MONTHS  |    CHILDREN   |     ADULTS    |     PAIRS")
print("-"*53)
fib(x)
```

```python
#rotate list by k position
def rotate_list(l1,k):
  l2=l1.split()
  l3=[int(i) for i in l2]
  l3=l3[k:]+l3[:k]
  print("\nRotated list : ",l3,"\n")

#list to convert list to tuple
def list_tup(l):
  t=tuple(l)
  return t
  print("list to tuple : ",t)

# remove duplicates
def remove_dupli(a1):
  return list(set(a1))

# to solve the given quadratic function
def solve_func(l1):
  return [i**2-i for i in l1]

# to merge list
def sorted_merge(p,q):
  sort_p=sorted(p)
  sort_q=sorted(q)
  merged_list=sorted(sort_p+sort_q)
  print("\nThe new merged list : ",merged_list)

a=input("Enter numbers : ")
b=int(input("Enter the position to rotate : "))
rotate_list(a,b)

a1=a.split()

a1=[int(i) for i in a1]
print("\nList to tuple : ",list_tup(a1))
print("\nBack to list by removing duplicates : ",remove_dupli(a1))
c=input("\nEnter numbers : ")
c1=c.split()
c1=[int(i) for i in c1]
solve_func(c1)
print("\nFunction solved : ",solve_func(c1))
sorted_merge(remove_dupli(a1),solve_func(c1))
```

```python
import random

class Box:
    def __init__(self, length, breadth=None, height=None):
        self.length = length
        self.breadth = breadth if breadth else length
        self.height = height if height else length
        self.area = self.calculate_area()
        self.volume = self.calculate_volume()

    def calculate_area(self):
        return 2 * (self.length * self.breadth + self.breadth * self.height + self.height * self.length)

    def calculate_volume(self):
        return self.length * self.breadth * self.height

# Number of boxes
N = int(input("Enter the number of boxes: "))
boxes = []

# Generate random measurements for the boxes and store them in the list
for _ in range(N):
    length = random.randint(1, 10)
    breadth = random.randint(1, 10)
    height = random.randint(1, 10)
    box = Box(length, breadth, height)
    boxes.append(box)

# Find the box with the maximum volume to area ratio
max_ratio_box = max(boxes, key=lambda box: box.volume / box.area)

# Print the details of the box with the maximum volume to area ratio
print("Details of the box with maximum volume to area ratio:")
print("Length:", max_ratio_box.length)
print("Breadth:", max_ratio_box.breadth)
print("Height:", max_ratio_box.height)
print("Area:", max_ratio_box.area)
print("Volume:", max_ratio_box.volume)

```
